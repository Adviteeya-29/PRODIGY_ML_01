```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
!unzip /content/housing_price_dataset.csv.zip
```

```python
df=pd.read_csv('housing_price_dataset.csv')
```

```python
df.head()
```

```python
#x contains all the features required(independent variables)
X=df[['SquareFeet', 'Bedrooms', 'Bathrooms']]
X
```

```python
# Y contains the target value(dependent variable)
Y=df['Price']
Y
```

```python
# here we have used 30% of the data for testing
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.3)
```

```python
model=LinearRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred= model.predict(X_test)
```

```python
mse= mean_squared_error(Y_test, Y_pred)
```

```python
#r2 determines the performance of the model
r2= r2_score(Y_test, Y_pred)
```

```python
#graph representing actual price and predicted price
plt.scatter(Y_test, Y_pred)
plt.xlabel('Actual prices')
plt.ylabel('Predicted prices')
plt.title('Actual prices vs Predicted prices')
plt.legend()
plt.show()
```

```python
print('mean squared error = ' , mse)
print('R square = ' , r2)
```
